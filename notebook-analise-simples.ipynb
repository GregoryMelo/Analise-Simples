### Importando as bibliotecas necessarias.
![Python](https://img.shields.io/badge/python-3670A0?style=for-the-badge&logo=python&logoColor=ffdd54)

```python
import pandas as pd # Importando a biblioteca "Pandas" para a manipulação da base de dados.
import matplotlib.pyplot as plt # Importando a biblioteca "Matplotlib" para a criação de graficos.
import seaborn as sns # Importando a biblioteca "Seaborn" para a estilização dos graficos.
```

### Lendo o arquivo excel que contem a base de dados.

```python
df = pd.read_excel('DataFrame.xlsx')
data = df.copy().dropna() # Fazendo uma copia dos dados da tabela e removendo valores nulos.
```

### Tratando a base de dados.

```python
data['Data'] = pd.to_datetime(data['Data']) # Convertendo a coluna data, sendo compativel com pandas.
data['Total'] = data['Quantidade'] * data['Preco_unitario'].astype(float) # Criando uma nova coluna com o total de cada item.
data['Produto'] = data['Produto'].str.strip() # Verificando se as espaçoes em brancos nas celulas

produtos_agrupados = data.groupby('Produto')['Quantidade'].sum().reset_index() # Agrupar por produto e somar as quantidades.
top_5_produtos = produtos_agrupados.sort_values(by='Quantidade', ascending=False).head(5) # Ordenar os produtos pela quantidade vendida em ordem decrescente
```

### Criação dos grafico.

Grafico de barras

```python
sns.set_style("white") # Definindo o estilo do Seaborn (opcional).
f, ax = plt.subplots(figsize=(20, 5)) # Criando e definindo o tamanho da figura do grafico.

ax.set_facecolor('white') # Definindo o fundo da área do gráfico como branco.
bars = sns.barplot(data=top_5_produtos, x='Produto', y='Quantidade',color="lightblue", ax=ax) # Criando o grafico.

# --- Adicionando rótulos e títulos ---
ax.bar_label(ax.containers[0], padding=0.3, fontsize=14)
ax.set_title("Top 5 Produtos Mais Vendidos", fontsize=24)
ax.set_xlabel("Produtos", fontsize=18)
ax.set_ylabel("Quantidade", fontsize=18)
ax.set_ylim((0, top_5_produtos['Quantidade'].max() + 5))


plt.show() # Mostrando o gráfico.
```
